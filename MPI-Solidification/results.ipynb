```python
PD = pd.read_csv("../profile_data.csv", header=0).to_numpy()
D1024 = PD[:7].transpose()
print(D1024)
D512 = PD[7:].transpose()
print(D512)
```

```python
X = np.log2(D1024[0])
T1 = D1024[2]
T5 = D512[2]
```

## Run Time

```python
plt.figure(figsize=(30,20))
plt.rc('xtick', labelsize=42)
plt.rc('ytick', labelsize=42)
plt.rc('font', family='serif', size=50)
plt.plot(X, T1,'o--', markersize=25)
plt.plot(X, T5,'s--', markersize=25)
plt.xlabel("No. of processors ($log_2(N)$)")
plt.ylabel("Run Time (seconds)")
plt.legend(["1024x1024", "512x512"])
```

```python
plt.figure(figsize=(30,20))
plt.rc('xtick', labelsize=42)
plt.rc('ytick', labelsize=42)
plt.rc('font', family='serif', size=50)
plt.plot(X, T1[0]/T1,'o--', markersize=30)
plt.plot(X, T5[0]/T5,'s--', markersize=30)
plt.xlabel("No. of processors ($log_2(N)$)")
plt.ylabel("Parallel Speedup S(N) = T(1)/T(N)")
plt.legend(["1024x1024", "512x512"])
```

```python
plt.figure(figsize=(30,20))
plt.rc('xtick', labelsize=42)
plt.rc('ytick', labelsize=42)
plt.rc('font', family='serif', size=50)
plt.plot(X, (T1[0]/T1)/(np.power(2,X)),'o--', markersize=30)
plt.plot(X, (T5[0]/T5)/(np.power(2,X)),'s--', markersize=30)
plt.xlabel("No. of processors ($log_2(N)$)")
plt.ylabel("Efficiency S(N)/N")
plt.legend(["1024x1024", "512x512"])
```
